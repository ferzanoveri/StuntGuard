```python
def calculate_energy(row):
    age_months = row['Age']
    body_weight = row['Body Weight']
    body_length = row['Body Length']

    if row['Gender'] == 1:
        if age_months <= 36:  # 0-3 years
            return (0.167 * body_weight) + (15.174 * body_length) - 617.6
        elif age_months <= 120:  # >3-10 years
            return (19.49 * body_weight) + (1.303 * body_length) + 414.9
    elif row['Gender'] == 0:
        if age_months <= 36:  # 0-3 years
            return (16.252 * body_weight) + (10.232 * body_length) - 413.5
        elif age_months <= 120:  # >3-10 years
            return (16.969 * body_weight) + (1.618 * body_length) + 371.2
    return None
```

```python
def calculate_protein(energy):
    if energy is None:
        return None

    protein_calories = 0.15 * energy  # 15% of energy needs from protein
    protein_grams = protein_calories / 4  # 1 gram of protein equals 4 calories
    return protein_grams
```

```python
# Apply the functions calculate energy & protein
df['Energy_kal'] = df.apply(calculate_energy, axis=1)
df['Protein_g'] = df['Energy_kal'].apply(calculate_protein)
df
```

```python
# Prepare food dataset
food_df_new = pd.read_csv('drive/MyDrive/Capstone/Dataset/Makanan_Dataset.csv')
food_df = food_df_new.drop(['Jumlah Porsi', 'Takaran Porsi'], axis=1)
food_df = food_df.loc[:, ['Kalori (kal)', 'Protein (g)']]
food_df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

# Standardize the food dataset
scaler = StandardScaler()
food_df_scaled = scaler.fit_transform(food_df.to_numpy())
```

```python
from sklearn.neighbors import NearestNeighbors

# Train the KNN model
model = NearestNeighbors(n_neighbors=50)
model.fit(food_df_scaled)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer

# Create the pipeline
transformer = FunctionTransformer(model.kneighbors, kw_args={'return_distance': False})
pipeline = Pipeline([('std_scaler', scaler), ('NN', transformer)])
pipeline.set_params(NN__kw_args={'n_neighbors': 8, 'return_distance': False})
```

```python
import numpy as np
# Function to recommend foods using KNN and a greedy algorithm
def recommend_foods_greedy_knn(child, food_df, model, scaler, n_recommendations=10):
    # Extract energy and protein needs from the child data
    energy = child['Energy_kal']
    protein = child['Protein_g']

    # Initialize remaining energy and protein
    remaining_energy = energy
    remaining_protein = protein

    recommendations = []

    # Calculate the initial percentage to cover
    percentages = [0.4, 0.5, 0.5, 0.5, 1.0]

    for pct in percentages:
        if remaining_energy <= 0 and remaining_protein <= 0:
            break

        # Calculate the target values
        target_energy = remaining_energy * pct
        target_protein = remaining_protein * pct

        # Find the nearest neighbors
        input_data = np.array([[target_protein, target_energy]])
        input_data_scaled = scaler.transform(input_data)
        distances, indices = model.kneighbors(input_data_scaled, n_neighbors=5)

        # Select the food item that minimizes the difference
        best_food = None
        best_diff = float('inf')

        for idx in indices[0]:
            food_item = food_df.iloc[idx]
            energy_diff = abs(food_item['Kalori (kal)'] - target_energy)
            protein_diff = abs(food_item['Protein (g)'] - target_protein)
            total_diff = energy_diff + protein_diff

            if total_diff < best_diff:
                best_diff = total_diff
                best_food = food_item

        if best_food is not None:
            recommendations.append(best_food.to_dict())
            remaining_energy -= best_food['Kalori (kal)']
            remaining_protein -= best_food['Protein (g)']
            food_df = food_df.drop(best_food.name)

    return recommendations
```

```python
# Select one child from the dataset and print results
example_idx = df.index[20]
example_child = df.loc[example_idx]

print(f"Example child (Age: {example_child['Age']} month, Gender: {'Male' if example_child['Gender'] == 1 else 'Female'}):")
print(f"Energy needs: {example_child['Energy_kal']} kal")
print(f"Protein needs: {example_child['Protein_g']} g")

# Get recommended foods using the KNN and greedy algorithm
recommended_foods = recommend_foods_greedy_knn(example_child, food_df_new, model, scaler)

# Create a DataFrame from recommended foods
recommended_foods_df = pd.DataFrame(recommended_foods)

# Print the recommended foods in table format
print("\nRecommended foods:")
print(recommended_foods_df)

# Calculate total recommended calories and protein
total_energy = recommended_foods_df['Kalori (kal)'].sum()
total_protein = recommended_foods_df['Protein (g)'].sum()

print(f"\nTotal calories from all recommended foods: {total_energy} kal")
print(f"Total protein from all recommended foods: {total_protein} g")
```
