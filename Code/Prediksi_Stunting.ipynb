```python
data = df.drop_duplicates()
data
```

```python
# label encode the Gender column using .loc
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
# label encode the Breastfeeding column
le = LabelEncoder()
data['Breastfeeding'] = le.fit_transform(data['Breastfeeding'])
```

```python
# label encode the Breastfeeding column
le = LabelEncoder()
data['Stunting'] = le.fit_transform(data['Stunting'])
```

```python
# Step 2: Preprocess the Data
# Assuming 'target' is the name of the target variable
X = data.drop(columns=['Stunting'])
y = data['Stunting']
```

```python
X
```

```python
y
```

```python
# split the train and test dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                       test_size=0.2,
                                       random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Inisialisasi model dengan hyperparameter yang bisa dioptimalkan
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the classifier
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", classification_rep)
```

```python
# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the results
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
```

```python
# Data baru untuk prediksi
new_data = [[1,16,2.8,49,7.7,49.0,0]]
# Melakukan prediksi label
y_pred = model.predict(new_data)
# Mencetak prediksi
print(f"Predicted label: {y_pred}")
```
